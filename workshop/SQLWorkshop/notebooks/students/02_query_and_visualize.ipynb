**Goal:** Join the fact table to dimensions, aggregate by year/state/gender, and create a bar chart.

1. Run the query below (totals by year, state, gender).
2. From the results, click **+ Add visualization** and create a **Bar chart**: X = state, Y = sum of total_claim_amount.
3. Save the query with a descriptive name (e.g. "Claim totals by state and year").

```python
USE CATALOG healthcare_dev;
USE SCHEMA cms;
```

```python
-- Totals by year, state, and gender (join fact to dim_beneficiary)
SELECT
  YEAR(f.claim_start_date) AS year,
  b.state,
  b.gender,
  SUM(f.claim_payment_amount) AS total_claim_amount,
  COUNT(f.claim_id) AS claim_count
FROM fact_patient_claims f
JOIN dim_beneficiary b ON f.beneficiary_key = b.beneficiary_key
WHERE f.claim_start_date IS NOT NULL
GROUP BY 1, 2, 3
ORDER BY year, state, gender;
```

```python
-- Simpler: totals by state only (for a bar chart X=state, Y=total_claim_amount)
SELECT
  b.state,
  SUM(f.claim_payment_amount) AS total_claim_amount,
  COUNT(f.claim_id) AS claim_count
FROM fact_patient_claims f
JOIN dim_beneficiary b ON f.beneficiary_key = b.beneficiary_key
WHERE b.state IS NOT NULL
GROUP BY b.state
ORDER BY total_claim_amount DESC;
```
