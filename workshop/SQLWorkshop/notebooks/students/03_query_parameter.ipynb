**Goal:** Add a **Year** parameter to your query and refresh results by changing the parameter.

1. In the SQL Editor, add a parameter (e.g. **Year** as integer or text).
2. Use it in the WHERE clause to filter by `YEAR(claim_start_date)`.
3. Run the query; change the parameter value and run again to see filtered results.

```python
USE CATALOG healthcare_dev;
USE SCHEMA cms;
```

```python
-- Parameter: set Year in the query settings or use a widget/variable
-- Example: bind parameter name 'Year' to value 2020, 2021, etc.
SELECT
  b.state,
  b.gender,
  SUM(f.claim_payment_amount) AS total_claim_amount,
  COUNT(f.claim_id) AS claim_count
FROM fact_patient_claims f
JOIN dim_beneficiary b ON f.beneficiary_key = b.beneficiary_key
WHERE YEAR(f.claim_start_date) = ${Year}
GROUP BY b.state, b.gender
ORDER BY total_claim_amount DESC;
```

**In Databricks SQL:** Create a query parameter named `Year` (integer). When you run the query, you will be prompted for the value, or set a default (e.g. 2020).
