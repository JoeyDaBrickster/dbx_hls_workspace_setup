# 05 – Dashboard

**Goal:** Create a dashboard from the reporting table **rprt_patient_claims**, add visualizations and filters.

## Steps
1. Go to **SQL → Dashboards** and create a **new dashboard**.
2. **Add panel** and choose **New query** (or an existing query).
3. Write a query (or use the assistant) to build **Claim amount by year and gender** (trend or bar chart).
4. Add a second panel: **Claim amount by state** (bar chart).
5. Add a **Gender** filter (and optionally State or Year) to the dashboard.
6. **Publish** and interact with the filters.

Use the queries below as a starting point.

```python
USE CATALOG healthcare_dev;
USE SCHEMA cms;
```

```python
-- Panel 1: Claim amount by year and gender (use as trend/line or bar)
SELECT
  YEAR(claim_start_date) AS year,
  gender,
  SUM(claim_payment_amount) AS total_claim_amount,
  COUNT(*) AS claim_count
FROM rprt_patient_claims
WHERE claim_start_date IS NOT NULL
GROUP BY 1, 2
ORDER BY 1, 2;
```

```python
-- Panel 2: Claim amount by state (bar chart)
SELECT
  state,
  SUM(claim_payment_amount) AS total_claim_amount,
  COUNT(*) AS claim_count
FROM rprt_patient_claims
WHERE state IS NOT NULL
GROUP BY state
ORDER BY total_claim_amount DESC;
```
