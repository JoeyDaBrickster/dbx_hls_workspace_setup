# 06 – Genie

**Goal:** Use the workshop Genie Space to answer natural-language questions and inspect how "local" and other conventions are applied.

## Steps
1. Open **Genie** and connect to the Space configured by your instructor (see **docs/GENIE_SPACE_CONFIG.md**).
2. Ask: **"Claims for patients who saw a local provider."**
3. Inspect how **"local"** is defined (e.g. NY). Review the generated SQL and any auto-visualizations.
4. Ask follow-up questions, for example:
   - Total claim amount by state.
   - How many claims in New York in 2020?
   - Claim amounts by year and gender.
5. Compare results with your earlier SQL and dashboard.

## Behavior rules (summary)
- **Local** = New York (NY).
- State abbreviations = two-letter (NY, CA, TX, …).
- *_flag fields = boolean.
- Default claim amount/date fields are used when not specified.
