# RESIGNAL Statement

**Applies to:** Databricks SQL, Databricks Runtime 16.3+

Re-raises the condition that was handled by the current condition handler. Use **only inside a condition handler**. No parameters. Use it after logging or side effects so the caller still sees the original error.

```python
USE CATALOG healthcare_dev;
USE SCHEMA gold;
```

## Example: Log DIVIDE_BY_ZERO with GET DIAGNOSTICS, then RESIGNAL

```python
BEGIN
  DECLARE EXIT HANDLER FOR DIVIDE_BY_ZERO
  BEGIN
    DECLARE cond STRING;
    DECLARE msg STRING;
    GET DIAGNOSTICS CONDITION 1
      cond = CONDITION_IDENTIFIER,
      msg  = MESSAGE_TEXT;
    SELECT CONCAT('Handler logged: ', cond, ' - ', msg) AS handler_log;
    RESIGNAL;
  END;
  SELECT 10 / 0;
END
```
