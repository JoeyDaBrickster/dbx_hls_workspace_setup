# WHILE Statement

**Applies to:** Databricks SQL, Databricks Runtime 16.3+

Repeat statements **while** a condition is true.

```
[label:] WHILE cond DO stmt [...] END WHILE [label]
```

Use **label** with LEAVE or ITERATE.

````python
USE CATALOG healthcare_dev;
USE SCHEMA gold;
````

## Example: Sum odd numbers 1..10 (classic demo)

````python
BEGIN
  DECLARE sum_val INT DEFAULT 0;
  DECLARE num INT DEFAULT 0;
  sumNumbers: WHILE num < 10 DO
    SET num = num + 1;
    IF num % 2 = 0 THEN
      ITERATE sumNumbers;
    END IF;
    SET sum_val = sum_val + num;
  END WHILE sumNumbers;
  SELECT sum_val AS sum_odd_1_to_10;
END
````

## Example: Simulate batch count â€“ increment until we reach claim count from fact_patient_claims

````python
BEGIN
  DECLARE target BIGINT DEFAULT (SELECT LEAST((SELECT COUNT(*) FROM fact_cms_outpatient_claims) + (SELECT COUNT(*) FROM fact_cms_inpatient_claims), 10));
  DECLARE batch INT DEFAULT 0;
  WHILE batch < target DO
    SET batch = batch + 1;
  END WHILE;
  SELECT batch AS batches_simulated, target AS target_count;
END
````
