```python
import subprocess
import us

for state in us.states.STATES:
  subprocess.run(
      [
          "java",
          "-jar", "/Workspace/Users/joey.filichia@databricks.com/synthea/synthea-with-dependencies.jar",
          "-c", "/Workspace/Users/joey.filichia@databricks.com/synthea/synthea.properties",
          f"--exporter.baseDirectory=/Volumes/healthcare_dev/bronze/landing/{state.abbr}/",
          "-p", "10",
          state.name,
      ],
      check=True,  # raise exception if non-zero exit
  )
```
